```python
categories = recentGrads.Major_category.unique()
categories
```

```python
menCount, womenCount = [], []
for category in categories:
    menCount.append(recentGrads[recentGrads.Major_category == category].Men.sum())
    womenCount.append(recentGrads[recentGrads.Major_category == category].Women.sum())
print(menCount)
print(womenCount)
```

Comparing men and women amount in different major categories

```python
df = pd.DataFrame({"Men":menCount, "Women":womenCount}, index=categories)
df.plot.bar()
```
