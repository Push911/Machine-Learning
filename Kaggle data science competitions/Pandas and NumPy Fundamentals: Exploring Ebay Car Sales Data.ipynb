Mean price of cars created by brands defined above

```python
brandPriceMean = {}
price = 0
for brand in selectedBrands:
    meanPrice = autos.loc[autos.brand == brand, "price"].mean()
    brandPriceMean[brand] = meanPrice
brandPriceMean
```

```python
brandMileageMean = {}
for brand in selectedBrands:
    meanMileage = autos.loc[autos.brand == brand, "odometer_km"].mean()
    brandMileageMean[brand] = meanMileage
brandMileageMean
```

Converting date to uniform numeric data

```python
autos.date_crawled = pd.to_datetime(autos.date_crawled.str[:10]).dt.strftime("%Y%m%d").astype(int)
autos.ad_created = pd.to_datetime(autos.ad_created.str[:10]).dt.strftime("%Y%m%d").astype(int)
autos.last_seen = pd.to_datetime(autos.last_seen.str[:10]).dt.strftime("%Y%m%d").astype(int)
```

```python
for brand in autos.brand.value_counts().index:
    models = autos.loc[autos.brand == brand, "model"].value_counts(ascending=False).index
    print(brand, models[0] if len(models) > 0 else "None model is presented in data for this brand")
```

```python
autos.groupby(["odometer_km"]).mean()
```

```python
priceComparisonOfDamagedCars = {}
unrepairedModels = autos.loc[autos.unrepaired_damage == "yes", "model"]
for model in unrepairedModels:
    unrepairedModelMeanPrice = autos.loc[(autos.unrepaired_damage == "yes") & (autos.model == model), "price"].mean()
    repairedModelMeanPrice = autos.loc[(autos.unrepaired_damage == "no") & (autos.model == model), "price"].mean()
    priceComparisonOfDamagedCars[model] = unrepairedModelMeanPrice, repairedModelMeanPrice
priceComparisonOfDamagedCars
```

```python
cheaper = {}
for key in priceComparisonOfDamagedCars:
    value = priceComparisonOfDamagedCars[key]
    cheaper[key] = value[1] - value[0]

cheaper
```
